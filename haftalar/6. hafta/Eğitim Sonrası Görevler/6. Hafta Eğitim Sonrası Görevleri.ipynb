Daha sonra, içinde çift ve tek sayılar bulunduran bir liste tanımlayın ve liste üzerinde gezinerek ekrana sadece çift sayıları bastırın.

```python
def cift_mi(x):
    for i in x:
        if i % 2 == 0:
            print(i)
        else:
            raise ValueError("tek sayi")
cift_mi([2,22,222,2222,22222,44,67,])

        
```

# Problem 2

2 YetGen lideri ders sırasında canları sıkılınca oyun oynamak isterler. Ancak hangi oyunu oynayacaklarına karar veremezler ve son olarak kendi oyunlarını yazmaya karar verirler.
Liderlerimizin kodlamak istediği oyun kuralları şöyledir:
- Oyunda random olarak 0 ile 100 arasında bir sayı tutulur.
- Kullanıcıdan bilgisayarın tuttuğu sayıyı tahmin etmesi için bir sayı girmesini istenir.
- Eğer kullanıcının girdiği sayı bulunması gereken sayıdan küçükse  tahmininizi yükseltin, büyükse tahmininizi düşürün şeklinde bir çıktı verilir.
- Oyunun gerçeğe yakın olması için bilgisayarın çalışmasını simgelemek için 1 saniye bekleme süresi tanınır.
- Kullanıcı 5 hakkı vardır. 5 hakkı da kullanıp sayıyı bulamazsa oyunu kaybeder.

```python
import time
import random

rast_sayi = random.randint(0,100)
tahmin_hakki = 5

while True:

    tahmin = int(input("Tahmin girin: "))
    if tahmin < rast_sayi:
        print("Dogrulaniyor..")
        time.sleep(1)

        print(f"Daha yuksek bir sayi giriniz. Girdiginiz sayi: {tahmin}")
        tahmin_hakki -=1

        if tahmin_hakki == 0:
            print("Ne yazikki tahmin hakkiniz kalmadi.")
            print(f"Aranan sayi: {rast_sayi}")
    elif tahmin > rast_sayi:
        print("Dogrulaniyor..")
        time.sleep(1)

        print(f"Daha dusuk bir sayi giriniz. Girdiginiz sayi: {tahmin}")
        tahmin_hakki -=1

        if tahmin_hakki == 0:
             print("Ne yazikki tahmin hakkiniz kalmadi.")
             print(f"Aranan sayi: {rast_sayi}")
    else: 
        print("Dogrulaniyor..")
        time.sleep(1)

        print(f"Tebrikler! Dogru tahmin. Sayi: {rast_sayi}")
        break
```
