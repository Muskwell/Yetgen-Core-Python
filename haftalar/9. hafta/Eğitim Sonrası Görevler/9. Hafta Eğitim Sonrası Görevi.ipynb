# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random

class Kumanda:
    def __init__(self):
        self.tv_durum = False
        self.ses = 10
        self.kanallar = ["TRT", "TCL", "DMAX", "NOW", "TRT Belgesel", "Kanal D", "Show TV", "Star TV"]
        self.current_channel = None

    def tv_ac(self):
        if not self.tv_durum:
            self.tv_durum = True
            print("Televizyon açıldı.")
        else:
            print("Televizyon zaten açık.")

    def tv_kapat(self):
        if self.tv_durum:
            self.tv_durum = False
            print("Televizyon kapatıldı.")
        else:
            print("Televizyon zaten kapalı.")

    def ses_ayarlari(self):
        print(f"Mevcut ses seviyesi: {self.ses}")
        yeni_ses = input("Yeni ses seviyesini girin (0-100): ")
        if yeni_ses.isdigit():
            yeni_ses = int(yeni_ses)
            if 0 <= yeni_ses <= 100:
                self.ses = yeni_ses
                print(f"Ses seviyesi ayarlandı: {self.ses}")
            else:
                print("Ses seviyesi 0-100 arasında olmalıdır.")
        else:
            print("Geçersiz ses seviyesi.")

    def kanal_ekle(self):
        kanal = input("Eklemek istediğiniz kanal adını girin: ")
        if kanal not in self.kanallar:
            self.kanallar.append(kanal)
            print(f"{kanal} kanalı eklendi.")
        else:
            print("Kanal zaten mevcut.")

    def kanal_sayisini_ogrenme(self):
        print(f"Toplam kanal sayısı: {len(self.kanallar)}")

    def rastgele_kanala_gec(self):
        if self.tv_durum:
            if self.kanallar:
                self.current_channel = random.choice(self.kanallar)
                print(f"Şu anda {self.current_channel} kanalındasınız.")
            else:
                print("Eklenmiş kanal yok.")
        else:
            print("Televizyon kapalı. Kanala geçiş yapamazsınız.")

    def tv_bilgileri(self):
        durumu = "Açık" if self.tv_durum else "Kapalı"
        print(f"Televizyon Durumu: {durumu}")
        print(f"Ses Seviyesi: {self.ses}")
        print(f"Toplam Kanal Sayısı: {len(self.kanallar)}")
        if self.current_channel:
            print(f"Şu anki Kanal: {self.current_channel}")
        else:
            print("Şu anda bir kanalda değilsiniz.")

def ana_program():
    kumanda = Kumanda()
    
    while True:
        print("\nKumanda Menüsü:")
        print("1. TV Aç")
        print("2. TV Kapat")
        print("3. Ses Ayarları")
        print("4. Kanal Ekle")
        print("5. Kanal Sayısını Öğrenme")
        print("6. Rastgele Kanala Geçme")
        print("7. Televizyon Bilgileri")
        print("q. Çıkış")
        
        secim = input("Bir seçenek girin: ")

        if secim == "1":
            kumanda.tv_ac()
        elif secim == "2":
            kumanda.tv_kapat()
        elif secim == "3":
            kumanda.ses_ayarlari()
        elif secim == "4":
            kumanda.kanal_ekle()
        elif secim == "5":
            kumanda.kanal_sayisini_ogrenme()
        elif secim == "6":
            kumanda.rastgele_kanala_gec()
        elif secim == "7":
            kumanda.tv_bilgileri()
        elif secim == "q":
            print("Programdan çıkılıyor...")
            break
        else:
            print("Geçersiz seçenek, lütfen tekrar deneyin.")

ana_program()
```
